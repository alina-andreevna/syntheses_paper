### Analyzing with 2 output regs

```python
df_reg2_ff = df.loc[df['Output_Registers_Number'] == 2].loc[df['Parameter_Name'] == 'Flip_Flops']
plot = sns.barplot(x="Resource_sharing_value", y="Parameter_Value", hue="Retiming_value", data=df_reg2_ff);
plot.set(ylabel = 'FF count')
plot.legend(bbox_to_anchor= (1.03, 0.6), title='Retiming_value')
fig = plot.get_figure()
fig.savefig("reg2_ff.png", dpi=300, bbox_inches = "tight")
```

```python
df_reg2_lookup = df.loc[df['Output_Registers_Number'] == 2].loc[df['Parameter_Name'] == 'Look-Up Tables']
plot = sns.barplot(x="Resource_sharing_value", y="Parameter_Value", hue="Retiming_value", data=df_reg2_lookup);
plot.set(ylabel = 'LUT count')
plot.legend(bbox_to_anchor= (1.03, 0.6), title='Retiming_value')
fig = plot.get_figure()
fig.savefig("reg2_lookup.png", dpi=300, bbox_inches = "tight")
```

```python
df_reg2_wns = df.loc[df['Output_Registers_Number'] == 2].loc[df['Parameter_Name'] == 'Worst Negative Slack']
plot = sns.barplot(x="Resource_sharing_value", y="Parameter_Value", hue="Retiming_value", data=df_reg2_wns);
plot.set(ylabel = 'WNS value, ns')
plot.legend(bbox_to_anchor= (1.03, 0.6), title='Retiming_value')
fig = plot.get_figure()
fig.savefig("reg2_wns.png", dpi=300, bbox_inches = "tight")
```

###Compare between regs

Comparasion between 2 and 4 output regs, so drop 3 and 5 regs:

```python
df_2_4 = df.drop(df[(df.Output_Registers_Number == 3) | (df.Output_Registers_Number == 5)].index)
```

#### Retiming on

```python
df_rt_on_ff = df_2_4.loc[df_2_4['Retiming_value'] == 'On'].loc[df_2_4['Parameter_Name'] == 'Flip_Flops']
plot = sns.barplot(x="Resource_sharing_value", y="Parameter_Value", hue="Output_Registers_Number", data=df_rt_on_ff);
plot.set(ylabel = 'FF count')
plot.legend(bbox_to_anchor= (1.03, 0.6), title='Output_Registers_Number')
fig = plot.get_figure()
fig.savefig("retiming_on_ff.png", dpi=300, bbox_inches = "tight")
```

```python
df_rt_on_lookup = df_2_4.loc[df_2_4['Retiming_value'] == 'On'].loc[df_2_4['Parameter_Name'] == 'Look-Up Tables']
plot = sns.barplot(x="Resource_sharing_value", y="Parameter_Value", hue="Output_Registers_Number", data=df_rt_on_lookup)
plot.set(ylabel = 'LUT count')
plot.legend(bbox_to_anchor= (1.03, 0.6), title='Output_Registers_Number')
fig = plot.get_figure()
fig.savefig("retiming_on_lookup.png", dpi=300, bbox_inches = "tight")
```

```python
df_rt_on_wns = df_2_4.loc[df_2_4['Retiming_value'] == 'On'].loc[df_2_4['Parameter_Name'] == 'Worst Negative Slack']
plot = sns.barplot(x="Resource_sharing_value", y="Parameter_Value", hue="Output_Registers_Number", data=df_rt_on_wns)
plot.set(ylabel = 'WNS value, ns')
plot.legend(bbox_to_anchor= (1.03, 0.6), title='Output_Registers_Number')
fig = plot.get_figure()
fig.savefig("retiming_on_wns.png", dpi=300, bbox_inches = "tight")
```
